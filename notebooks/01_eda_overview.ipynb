```python
# Импорты и настройки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Настройки отображения
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 5)
plt.rcParams["figure.dpi"] = 100

INTERIM_DIR = Path("../data/interim")
RAW_DIR = Path("../data/raw")

print("Готово. Пути настроены.")
```

```python
# Загрузка данных
subway = pd.read_parquet(INTERIM_DIR / "subway_hourly.parquet")
bus = pd.read_parquet(INTERIM_DIR / "bus_hourly.parquet")
alerts = pd.read_parquet(RAW_DIR / "service_alerts.parquet")

print(f"Subway: {len(subway):,} строк, {subway['station_complex_id'].nunique()} станций")
print(f"Bus:    {len(bus):,} строк, {bus['bus_route'].nunique()} маршрутов")
print(f"Alerts: {len(alerts):,} строк")
print()
print("Период subway:", subway["transit_timestamp"].min(), "—", subway["transit_timestamp"].max())
print("Период bus:   ", bus["transit_timestamp"].min(), "—", bus["transit_timestamp"].max())
```
